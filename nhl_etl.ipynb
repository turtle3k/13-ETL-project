```python
import pandas as pd
from sqlalchemy import create_engine
import datetime as dt
```

### Store CSV into DataFrame

```python
csv_file = "Resources/Teams.csv"
team_data_df = pd.read_csv(csv_file)
team_data_df.head()
```

### Create new data with select columns

```python
new_team_data_df = team_data_df[['year', 'rank', 'name', 'confID', 'divID']].copy()
new_team_data_df.head()
```

```python
new_team_data_df = new_team_data_df.groupby(['year']).get_group((2011))
```

```python

new_team_data_df.head()
```

```python
new_team_data_df = new_team_data_df.rename(columns={"name": "team1", "year": "season",
                                                    "rank": "division_rank", "confID": "conference",
                                                    "divID": "division"
                                              })
#df.rename(columns={"A": "a", "B": "c"})
new_team_data_df.head()
```

```python
# Clean season to match attendance Season format
new_team_data_df['season'].apply(str)
new_team_data_df['season'] = new_team_data_df['season'].apply(str)
```

```python
new_team_data_df['season'] = new_team_data_df['season'].replace({'2011':"2010-11"})
```

```python
new_team_data_df
```

```python
# Replace spaces in city names, so we can split city & team
#new_team_data_df.Team1.replace(['Los Angeles', 'New Jersey', 'New York', 'San Jose', 'St. Louis', 'Tampa Bay'], 
                              #['Los_Angeles', 'New_Jersey', 'New_York', 'San_Jose', 'St._Louis', 'Tampa_Bay'], inplace=True)
#new_team_data_df['Season'] = new_team_data_df['Season'].replace({'2011':"2010-11"})
new_team_data_df['team1'] = new_team_data_df['team1'].replace({'Los Angeles Kings':'Los_Angeles Kings',
                                                              'New Jersey Devils':'New_Jersey Devils',
                                                               'New York Islanders': 'NY Islanders',
                                                               'New York Rangers': 'NY Rangers',
                                                               'San Jose Sharks': 'San_Jose Sharks',
                                                               'St. Louis Blues': 'St._Louis Blues',
                                                               'Tampa Bay Lightning': 'Tampa_Bay Lightning'})
                                                                                                                          
new_team_data_df
```

```python
temp_team_data_df = new_team_data_df["team1"].str.split(" ",n=1,expand=True)
new_team_data_df["city"]= temp_team_data_df[0]
new_team_data_df["team"]= temp_team_data_df[1]
new_team_data_df.drop(columns = {"team1"}, inplace = True)
```

```python
#new_team_data_df = new_team_data_df.set_index('Team')
new_team_data_df
```

### Store second CSV data file into a DataFrame

```python
csv_file = "Resources/NHLAttendance.csv"
attendance_df = pd.read_csv(csv_file)
attendance_df.head()
```

```python
attendance_df.columns.values
```

### Clean DataFrame

```python
new_attendance_df = attendance_df[["SEASON", "RANK", "TEAM", " HOME ATTENDANCE ", " ROAD ATTENDANCE ", " TOTAL ATTENDANCE "]].copy()
new_attendance_df.head()
```

```python
new_attendance_df = new_attendance_df.groupby(['SEASON']).get_group(('2010-11'))
new_attendance_df
```

```python
new_attendance_df = new_attendance_df.rename(columns={"SEASON": "season", "RANK": "rank",
                                                    "TEAM": "city", " HOME ATTENDANCE ": "home_attendance",
                                                    " ROAD ATTENDANCE ": "road_attendance",
                                                    " TOTAL ATTENDANCE ": "total_attendance"  
                                              })
```

```python
#Rename cities in the attendance df to match those in the teams df
new_attendance_df['city'] = new_attendance_df['city'].replace({'Los Angeles':'Los_Angeles',
                                                              'New Jersey':'New_Jersey',
                                                               'New York': 'New_York',
                                                               'San Jose': 'San_Jose',
                                                               'St. Louis': 'St._Louis',
                                                               'Tampa Bay': 'Tampa_Bay'})
                   
```

```python
new_attendance_df
```

### Connect to local database

```python
rds_connection_string = "postgres:PGD@tabas3pw@localhost:5432/NHL2011_DB"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

### Check for tables

### Use pandas to load TEAM csv converted DataFrame into database

```python
new_team_data_df.to_sql(name='team_table', con=engine, if_exists='append', index=False)
```

### Use pandas to load ATTENDANCE csv converted DataFrame into database

```python
new_attendance_df.to_sql(name='attendance_table', con=engine, if_exists='append', index=False)
```
